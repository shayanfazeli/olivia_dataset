```python
import numpy
import pandas
import os
import sys
sys.path.append(os.path.abspath('../../'))
from epiweeks import Week
```

# COVID_5 Dashboard

#### Case Counts

```python
weekly_data_counts = pandas.read_csv(os.path.join('COVID19_5', 'Weekly_Data_Counts.csv'), skiprows=2)
weekly_data_counts.dropna(inplace=True)
weekly_data_counts['MMWR-YEAR'] = weekly_data_counts['WEEK_NUMBER'].apply(lambda x: x.split('-')[0])
weekly_data_counts['MMWR-WEEK'] = weekly_data_counts['WEEK_NUMBER'].apply(lambda x: x.split('-')[1])
weekly_data_counts['MMWR-DATE'] = weekly_data_counts.apply(lambda x: str(Week.fromstring(f"{x['MMWR-YEAR']}" + f"{int(x['MMWR-WEEK']):02d}").enddate()), axis=1)
```

#### Case Percentages

```python
weekly_data_percents = pandas.read_csv(os.path.join('COVID19_5', 'Weekly_Data_Percent.csv'), skiprows=2)
weekly_data_percents.dropna(inplace=True)
weekly_data_percents['MMWR-YEAR'] = weekly_data_percents['WEEK_NUMBER'].apply(lambda x: x.split('-')[0])
weekly_data_percents['MMWR-WEEK'] = weekly_data_percents['WEEK_NUMBER'].apply(lambda x: x.split('-')[1])
weekly_data_percents['MMWR-DATE'] = weekly_data_percents.apply(lambda x: str(Week.fromstring(f"{x['MMWR-YEAR']}" + f"{int(x['MMWR-WEEK']):02d}").enddate()), axis=1)
```
